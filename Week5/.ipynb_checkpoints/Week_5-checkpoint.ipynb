```python
# First this
import pandas as pd
import matplotlib.pyplot as plt
aData = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?lang=en&delimiter=Semicolon&CIVILSTAND=F&Tid=2008K1%2C2020K3', sep=';')
bData = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?lang=en&delimiter=Semicolon&OMR%C3%85DE=*&CIVILSTAND=U%2CTOT', sep=';')
cData = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?lang=en&delimiter=Semicolon&OMR%C3%85DE=101&Tid=*&CIVILSTAND=G', sep=';')
dData = pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?lang=en&delimiter=Semicolon&OMR%C3%85DE=000&ALDER=*&CIVILSTAND=U%2CG&Tid=2020K3', sep=';')
```

```python
# a) What is the change in pct of divorced danes from 2008 to 2020?

divorces_2008Q1 = aData.iloc[0,2]
divorces_2020Q3 = aData.iloc[1,2]
print('Change in pct from 2008Q1 to 2020Q3: ', (divorces_2020Q3 - divorces_2008Q1)/divorces_2008Q1 *100)
```

```python
# b) Which of the 5 biggest cities has the highest percentage of 'Never Married'?

rmThis = ['All Denmark', 'Region Hovedstaden', 'Region Sjælland', 'Region Syddanmark', 'Region Midtjylland', 'Region Nordjylland']

bDataTotal = bData[(~bData['OMRÅDE'].isin(rmThis)) & (bData['CIVILSTAND'] == 'Total')]

#We need to sort here in order to find the 5 biggest cities.
sort_for_size = bDataTotal.sort_values(by='INDHOLD', ascending=False)[:5]
bdDataCity = sort_for_size.iloc[:,0]

#We find the percent of unmarried people for each of the 5 biggest cities.
unmarried = []
for index, city in bdDataCity.iteritems():
    unmarried.append([city, bData.iloc[index-1]['INDHOLD']/bData.iloc[index]['INDHOLD']*100])

#We sort them again after percentage of unmarried people.
unmarried = sorted(unmarried, key=lambda x:x[1], reverse=True)
# for city in unmarried:
#     print(city)

print('Highest percent of unmarried people: ', unmarried[0])
```

```python
# c) Show a bar chart of changes in marrital status in Copenhagen from 2008 till now

cData.plot(kind='bar', title='Marritial Status', figsize=(10,5), x='TID', y='INDHOLD')
```

```python
# d) Show a bar chart of 'Married' and 'Never Married' for all ages in DK (Hint: 2 bars of different color)

dDataEdit = dData[(dData['ALDER'] != 'Total')]
#dDataEdit.plot(kind="bar", title='Marritial Status',figsize=(10,5), x='ALDER', y='INDHOLD')

never_married = dDataEdit[(dDataEdit['CIVILSTAND'] == 'Never married')].reset_index(drop=True)
married = dDataEdit[(dDataEdit['CIVILSTAND'] == 'Married/separated')].reset_index(drop=True)

plotData = pd.concat([never_married, married], axis=1, keys=['Never Married', 'Married'])
plotData.plot(kind='bar', figsize=(10,5), title='Marritial Status')


#label = never_married['ALDER'].values
#width = 0.35

#plt.bar(never_married['ALDER'], never_married['INDHOLD'])
#plt.bar(married['ALDER'], married['INDHOLD'])

#figure = plt.figure(figsize=(10,5))
#plt.bar(never_married['ALDER'], never_married['INDHOLD'], width, label='Never Married', color='blue')
#plt.bar(married['ALDER'], married['INDHOLD'], width, label='Married', color='red')
#plt.ylabel('Amount of people')
#plt.xlabel('Age of people')
#plt.title('Marritial status')
#plt.xticks(label, rotation=90)
#ax.set_xticklabels(label)
#plt.legend()
#plt.show()
```
