```python
#Exercise 1.10

# 10. Show a bar chart of distribution of study progression on x-axis and number of students in each category on y-axis. (e.g. make 10 categories from 0-100%)
import my_modules.ex1_8Utils as utils8
import matplotlib.pyplot as plt
studentsFromCSV = utils8.readStudentDataFromCSV("./my_resources/students.csv")
sortedList = []
for student in studentsFromCSV:
    sortedList.append([student.name, student.get_progression()])
name = []
progression = []
sortedList.sort(key=lambda x: x[1], reverse=True)
for item in sortedList:
    name.append(item[0])
    progression.append(item[1])
plt.figure()
plt.bar(name, progression)
plt.title("Bar chart", fontsize=24)
plt.xlabel("Names", fontsize=14)
plt.xticks(rotation=90)
plt.ylabel("ECTS procent", fontsize=14)
plt.show(block=True)
```
